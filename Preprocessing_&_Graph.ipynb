```python
# importing the required libraries
import string
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file = open(r"input_file.txt","rt") # open the file in the jupyter notebook environment
```

```python
text = file.read()  # read the file
```

```python
type(text) #datatype of text variable
```

```python
text   #data inside text variable
```

```python
# splitting the text into list of words  by white space
words = text.split() 
# remove punctuation from each word

table = str.maketrans(' ', ' ', string.punctuation)
stripped = [w.translate(table) for w in words]
print(stripped[:100])
```

```python
# convert to lower case
stripped = [word.lower() for word in stripped]
print(stripped[:100])
```

```python
type(stripped) #datatype of stripped variable
```

```python
# count of words starting with http or www
count =0
for word in stripped:
    if(word.startswith('http') or word.startswith('www')):
        count=count + 1
print(count)
```

```python
# removing the words from stripped variable list which starts with http or www
for word in stripped:
    if(word.startswith('http') or word.startswith('www')):
        stripped.remove(word)
```

```python
# count of words starting with http or www
count =0
for word in stripped:
    if(word.startswith('http') or word.startswith('www')):
        count=count + 1
print(count)
```

```python
listToStr = ' '.join([str(elem) for elem in stripped]) # join the list of words into a single string
```

```python
listToStr 
```

```python
file1 = open(r"preprocessed.txt",'w') # open the preprocessed file (empty) in the jupyter notebook environment
```

```python
file1.write(listToStr)  # write the listToStr into the preprocessed.txt file
```

```python
file1.close()   #close the preprocessed.txt file 
```

```python
file2 = open(r"part-r-00000",'rt')  # open the output file which we received from mapreduce framework
```

```python
text1 = file2.read() # read the output file 
```

```python
# spliting the text1 string(from output file) line by line and storing the word, count as key value pair in dict
dict = {}
for line in text1.splitlines():
    a = line.split()
    dict[a[0]]=int(a[1])
```

```python
dict 
```

```python
# sorting the dict in desending order
sorted_dict={k: v for k, v in sorted(dict.items(), key=lambda item: item[1],reverse=True)} 
```

```python
del sorted_dict['gutenbergtm']
```

```python
sorted_dict
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
keys = list(sorted_dict. keys())      # taking the keys of the sorted_dict as a list 
keys = keys[30:60]                    # slicing the keys list and taking 30 keys
values = list(sorted_dict. values())  # taking the values of the sorted_dict as a list
values = values[30:60]                # slicing the values list and taking 30 keys
plt.bar(keys,values)                  # plotting the count(y-axis) vs word(x-axis)
plt.xticks(rotation=90)                # display the x-axis label vertically
```
